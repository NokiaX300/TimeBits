```python
# 查看模型列表
import ollama

models = ollama.list()
print("本地模型列表:")
for model in models['models']:
    print(model)
```

```python
# 运行基础对话
response = ollama.chat(
    model='qwen3:0.6b',
    messages=[{'role': 'user', 'content': '用中文解释强化学习'}]
)
print(response['message']['content'])
```

```python
# 流式输出
stream = ollama.chat(
    model='qwen3:0.6b',
    messages=[{'role': 'user', 'content': '写一个Python快速排序算法'}],
    stream=True
)

for chunk in stream:
    print(chunk['message']['content'], end='', flush=True)
```

```python
# 生成文本（非对话模式）
response = ollama.generate(
    model='qwen3:0.6b',
    prompt='将以下内容翻译成英文：人工智能正在改变世界',
    options={'temperature': 0.5}
)
print(response['response'])
```

```python
# 获取嵌入向量
embeddings = ollama.embeddings(
    model='qwen3:0.6b',
    prompt='量子计算的基本原理'
)
print(f"向量维度: {len(embeddings['embedding'])}")
print(embeddings)
```

```python
from threading import Thread
import time
import ollama

def run_model(model_name, question):
    start = time.time()
    print(f"[{time.strftime('%H:%M:%S')}] {model_name} 开始处理...")
    
    response = ollama.generate(model=model_name, prompt=question)
    
    end = time.time()
    print(f"[{time.strftime('%H:%M:%S')}] {model_name} 回答生成完毕 | 耗时: {end-start:.2f}s\n"
          f"答案片段: {response['response'][:50]}...")

# 启动两个线程
Thread(target=run_model, args=('qwen3:0.6b', '量子计算原理')).start()
Thread(target=run_model, args=('qwen3:14b', '区块链技术详解')).start()
```
